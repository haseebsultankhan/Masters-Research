# SEPSIS DATA SET

1: Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
# Adjusting column and row numbers to be shown
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

2: Converting different .psv files into a single file with patient_ID

```python
import glob

path = r'/home/faisal/Desktop/sepsis data set/' # use your path
all_files = glob.glob(path + "/*.psv")

li = []

for i,filename in enumerate (all_files):
    
    df = pd.read_csv(filename,'|',index_col=None, header=0).assign(Patient_iD=i)
    

    li.append(df)
```

3: Making a DataFrame

```python
data = pd.concat(li, axis=0, ignore_index=True)
df = data.copy()
```
